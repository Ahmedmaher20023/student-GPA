```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student_names = np.array(["Ali", "Ahmed", "Youssef", "sherif", "Joe", "Omar", "Rammi", "Hana", "Ziad", "Abdallah"])
```

```python
subjects_scores = np.random.randint(0, 100, (10, 5))
```

```python
total_degrees = np.sum(subjects_scores, axis=1)
percentages = total_degrees / subjects_scores.shape[1]
```

```python
def get_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    elif percentage >= 50:
        return 'E'
    else:
        return 'F'
```

```python
grades = np.array([get_grade(perc) for perc in percentages])
```

```python
num_students = len(student_names)
max_degree = np.max(total_degrees)
min_degree = np.min(total_degrees)
average_total = np.mean(total_degrees)
total_sum = np.sum(total_degrees)
count_exceed_50 = np.sum(percentages > 50)
```

```python
print("Student Data Table:\n", pd.DataFrame({
    "Student Name": student_names,
    "Sub1": subjects_scores[:, 0],
    "Sub2": subjects_scores[:, 1],
    "Sub3": subjects_scores[:, 2],
    "Sub4": subjects_scores[:, 3],
    "Sub5": subjects_scores[:, 4],
    "Total Degree": total_degrees,
    "Percentage": percentages,
    "Grade": grades
}))
```

```python
print("\nSummary:")
print(f"Number of Students: {num_students}")
print(f"Maximum Degree: {max_degree}")
print(f"Minimum Degree: {min_degree}")
print(f"Average of Totals: {average_total}")
print(f"Total Sum of Degrees: {total_sum}")
print(f"Count Exceeding 50%: {count_exceed_50}")
```
